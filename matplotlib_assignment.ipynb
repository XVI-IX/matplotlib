```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# A plot comparing rainfall and crop yield using mock randomly generated data
crop_yield = np.random.randint(0, 1500, 12)
rainfall = np.random.randint(0, 2000, 12)

crop_yield_1 = np.random.randint(0, 1500, 12)
rainfall_1 = np.random.randint(0, 2000, 12)

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',
        'Oct', 'Nov', 'Dec']
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

axes[0].plot(months, rainfall, label='Rainfall(mm)')
axes[0].set_title('CROP YIELD AND RAINFALL, 2019').set_color('red')
axes[0].set_ylabel('Rainfall').set_color('blue')
axes[0].spines['right'].set_color('orange')
axes[0].spines['left'].set_color('blue')
for label in axes[0].get_yticklabels():
    label.set_color('blue')
new_axes = axes[0].twinx()
new_axes.plot(months, crop_yield,'orange', label='Crop Yield (mt)')
new_axes.set_ylabel('Crop Yield').set_color('orange')
for label in new_axes.get_yticklabels():
    label.set_color('orange')
axes[0].grid(True)
new_axes.legend(loc=4)
axes[0].legend(loc=3)

axes[1].plot(months, rainfall_1, label='Rainfall (mm)')
axes[1].set_title('CROP YIELD AND RAINFALL, 2020').set_color('red')
axes[1].set_ylabel('Rainfall').set_color('blue')
axes[0].spines['left'].set_color('blue')
axes[0].spines['right'].set_color('orange')
for label in axes[0].get_yticklabels():
    label.set_color('blue')
new_axes = axes[1].twinx()
new_axes.plot(months, crop_yield_1,'orange', label='Crop Yield (mt)')
new_axes.set_ylabel('Crop Yield').set_color('orange')
for label in new_axes.get_yticklabels():
    label.set_color('orange')
axes[1].grid(True)
new_axes.legend(loc=3)
axes[1].legend(loc=0)
```

```python
#
years = ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018',
        '2019', '2020']
cost = np.random.randint(1, 200, 11)
profit = np.random.randint(0, 100, 11)
waste = np.random.randint(0, 100, 11)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 8))

ax[0].plot(years, cost, label='Cost($)')
ax[0].plot(years, waste,'orange', label='Waste')
ax[0].set_title('COST AND PROFIT').set_color('red')
ax[0].set_ylabel('Cost ($)').set_color('blue')
ax[0].grid(True)
ax[0].legend(loc=0)

ax[1].plot(years, cost, label='Cost(million $)')
ax[1].plot(years, waste,'orange', label='Waste')
ax[1].set_title('COST AND WASTE').set_color('red')
ax[1].set_ylabel('Cost ($)').set_color('blue')
ax[1].grid(True)
ax[1].legend(loc=0)
```
